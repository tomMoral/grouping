```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from SSA import SSA
```

```python
from artsig.artsig_article import Sig_Gen

gen = Sig_Gen()

sig, dec = gen.gen_set(N=1, h=1, seed=12455)
sig, dec = sig[0], dec[0]
plt.subplot(len(dec)+1, 1, 1)
plt.plot(sig)
for i, c in enumerate(dec):
    plt.subplot(len(dec)+1, 1, i+2)
    plt.plot(c)
```

```python
ssa = SSA(250)
sig_SSA = ssa.fit_transform(sig, tau_l=0.01)
print(sig_SSA.shape)

components = [ssa._compos_k(s, k) for k, s in enumerate(sig_SSA)]
```

```python
n_compos = min(len(components), 10)
for i, c in enumerate(components[:n_compos]):
    plt.subplot(n_compos, 1, i+1)
    plt.plot(c)
```

```python
from SSA.Grouping import Grouping_Algorithms
ga = Grouping_Algorithms(components, ssa.dic, ssa.vp)

ga._correlation()
plt.subplot(121)
plt.imshow(ga.c_mat, interpolation='none')
plt.subplot(122)
plt.imshow(ga.c_mat[:10, :10], interpolation='none')
```

```python
ga._w_correlation()
plt.subplot(121)
plt.imshow(ga.w_mat, interpolation='none')
plt.subplot(122)
plt.imshow(ga.w_mat[:10, :10], interpolation='none')
```

```python
for c in components:
    plt.plot(abs(np.fft.fft(c)))
plt.xlim((0, 100))
```
